```python
import pandas as pd
from pathlib import Path
from common.transforms import funding_to_daily_annualized, build_spread, log_return

RAW = Path('../data/raw')
PROCESSED = Path('../data/processed')

def pick_funding_path(symbol: str):
    cands = [RAW/f'binance_{symbol.lower()}_funding.csv', RAW/f'bybit_{symbol.lower()}_funding.csv']
    for p in cands:
        if p.exists():
            return p
    raise FileNotFoundError(f'funding file not found for {symbol}')

def load_core_inputs():
    eth_f = pd.read_csv(pick_funding_path('ETHUSDT'))
    btc_f = pd.read_csv(pick_funding_path('BTCUSDT'))
    eth_p = pd.read_csv(RAW/'binance_ethusdt_1d.csv')
    btc_p = pd.read_csv(RAW/'binance_btcusdt_1d.csv')
    st_path = PROCESSED/'eth_yield_panel.csv'
    if st_path.exists():
        st = pd.read_csv(st_path)
    else:
        st = pd.read_csv(RAW/'eth_staking_yield_daily.csv')
    return eth_f, btc_f, eth_p, btc_p, st
```
